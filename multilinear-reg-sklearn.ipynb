```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
boston = datasets.load_boston()
```

Above, we loaded the Boston Houses Dataset(a pre-installed scikit-learn dataset). Now, we will print the dictionary keys of the dataset.

```python
print(boston.keys())
```

Below we print out the Features of our selected Dataset, as well as the description.

```python
print(boston.feature_names, boston.DESCR)
```

```python
print(boston.data.shape)
```

Dependent Variable X and Independent Variable y

```python
X = boston.data
```

```python
y = boston.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=16)
```

```python
print(X_test.size)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions.size)
```

```python
print(predictions)
```

```python
plt.plot(X_test,y_test, '.',color='blue')
plt.plot(X_test,predictions,linewidth=1 ,color='orange')
plt.show()
#len(predictions)
```

```python
plt.plot(y_test,'.', color='blue')
plt.plot(predictions,linewidth=1 ,color='red')
plt.show()
#len(predictions)
```

```python

X_test=pd.DataFrame(X_test)
X_test.describe()
```

```python
plt.scatter(y_test, predictions)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
```

```python
plt.show()
```

```python
## plotting residual errors in training data 
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, 
            color = "green", s = 10, label = 'Train data') 
  
## plotting residual errors in test data 
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, 
            color = "blue", s = 10, label = 'Test data') 
  
## plotting line for zero residual error 
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2) 
  
## plotting legend 
plt.legend(loc = 'upper right') 
  
## plot title 
plt.title("Residual errors") 
print('Variance score: {}'.format(model.score(X_test, y_test)))
```

```python
print(model.score(X_test, y_test))
```
